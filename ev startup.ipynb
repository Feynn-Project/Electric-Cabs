```python
# putting independent variables as x and dependent variables as y
x= df[['AccelSec', 'Range_Km', 'TopSpeed_KmH', 'Efficiency_WhKm']]
y= df[['PriceEuro']]
```

```python
# Finding the Linear Regression using OLS mothod
import statsmodels.api as sm
x= sm. add_constant(x)
results= sm.OLS(y,x)
```

```python
model= results.fit()
model.summary()
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 365)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
```

```python
lr.fit(x_train, y_train)
pred= lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2= (r2_score, (y_test, pred))
print(r2*100)
```

around 78% of the dependent variables has been explained by the independent variables

```python
#from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test= train_test_split(x1, y1, test_size=0.2, random_state= 365) 
```

```python
from sklearn.linear_model import LogisticRegression
log= LogisticRegression()
```

```python
log.fit(x1_train, y1_train)
pred1= log.predict(x1_test)
pred1
```

```python
# confusion matrix of the regression
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y1_test, pred1)
cm
```

```python
# finding the accuracy score
from sklearn.metrics import accuracy_score
score= accuracy_score(y1_test, pred1)
score*100
```
